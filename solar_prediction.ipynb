```python
df1=df.drop(['Data','UNIXTime'],axis=1)
```

```python
df1['Time']=pd.to_datetime(df1["Time"])
```

```python
df1['hour']=df1["Time"].dt.hour
df1["minute"]=df1["Time"].dt.minute
df1["second"]=df1["Time"].dt.second

df1
```

```python
df1["TimeSunRise"]=pd.to_datetime(df1["TimeSunRise"])
df1["TimeSunSet"]=pd.to_datetime(df1["TimeSunSet"])
df1["SunRisehour"]=df1["TimeSunRise"].dt.hour
df1["SunRiseminite"]=df1["TimeSunRise"].dt.minute
df1["sunrisesecond"]=df1["TimeSunRise"].dt.second
df1["sunsethour"]=df1["TimeSunSet"].dt.hour
df1["sunsetminute"]=df1["TimeSunSet"].dt.minute
df1["sunsetsecond"]=df1["TimeSunSet"].dt.second
df1
```

```python
df1.drop(["TimeSunRise","TimeSunSet","Time"],axis=1)
```

```python
df1.dtypes
```

```python
y=df1['Radiation']
y
```

```python
x=df1.drop(['Radiation'],axis=1)
x
```

```python
x.drop(['Time','TimeSunRise','TimeSunSet'],axis=1,inplace=True)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error
a=mean_absolute_percentage_error(y_test,y_pred)
print(a)
```

```python
from sklearn.metrics import mean_absolute_error
print("mean absolute error",mean_absolute_error(y_test,y_pred))
```

```python
from numpy.core.fromnumeric import mean
from sklearn.metrics import mean_squared_error
b=mean_squared_error(y_test,y_pred)
b
```

```python
sq=np.sqrt(b)
sq
```
